```python
user_data_count = dict()

for line in rating_data:
    line = line.strip()
    UserID, _, _, _ = map(int, line.split('::'))

    if UserID in user_data_count:
        user_data_count[UserID] += 1
    else:
        user_data_count[UserID] = 1

# user_data_count = dict(sorted(user_data_count.items(), key=lambda x: x[1], reverse=True))
user_data_count
```

```python
userID_with_profession = dict()
    
for line in user_data:
    line = line.strip()
    UserID, _, _, profession, _ = line.split('::')
    userID_with_profession[int(UserID)] = int(profession)

userID_with_profession
```

```python
Occupation_count = { x : 0 for x in profession_dict.values()}

for UserID, count in user_data_count.items():
    
    profession = userID_with_profession[UserID]
    
    Occupation_count[profession_dict[profession]] += count


Occupation_count = sorted(Occupation_count.items(), key=lambda x: x[1], reverse=True)[:1]

print("the most active users (profession) based on the number of ratings ")
for  profession, count in Occupation_count:
    print(f"{profession}: {count}")
    
```

### Analyze the relationship between user demographic attributes (age, gender, occupation) and their movie preferences or rating patterns.

```python
# this is use for collection of user demographic_info


user_demographic_info_count = dict()

for line in user_data:
    line = line.strip()
    columns = line.split('::')
    columns[2] = int(columns[2])
    columns[3] = int(columns[3])
    columns.append(0)
    

    user_demographic_info_count[int(columns[0])] = columns[1:]

for line in rating_data: # 1::1193::5::978300760
    line = line.strip()
    UserID, _, Reating, _ = map(int, line.split('::'))
    
    user_demographic_info_count[UserID][4] += 1
    

print(user_demographic_info_count) # 1: ['F', 1, 10, '48067', 53],

     
```

```python
# movie watched by age group
age_group_count = {
                    "Under 18": 0,
                    "18-24": 0,
                    "25-34": 0,
                    "35-44": 0,
                    "45-49": 0,
                    "50-55": 0,
                    "56+": 0
                }

for list_ in user_demographic_info_count.values():
    age = list_[1]
    
    if age < 18:
        age_group_count["Under 18"] += list_[4]
    elif age < 25:
        age_group_count["18-24"] += list_[4]
    elif age < 35:
        age_group_count["25-34"] += list_[4]
    elif age < 45:
        age_group_count["35-44"] += list_[4]
    elif age < 50:
        age_group_count["45-49"] += list_[4]
    elif age < 56:
        age_group_count["50-55"] += list_[4]
    else:
        age_group_count["56+"] += list_[4]

age_group_count

print("Count of rating by age group:")
for age, count in age_group_count.items():
    print(f"Age {age}: {count}")
```

```python
#movie watched by gender group

movie_watch_count_gender = { i[0]: 0 for i in user_demographic_info_count.values()}

for list_ in user_demographic_info_count.values():
    gender = list_[0]
    movie_watch_count_gender[gender] += 1


movie_watch_count_gender # {'F': 1709, 'M': 4331} = 6040

print("Count of rating by gender:")
for gender, count in movie_watch_count_gender.items():
    print(f"Gender {gender}: {count}")
```

```python
# Calculate rating by occupation
 # 1: ['F', 1, 10, '48067', 53],

movie_watch_count_by_profession = dict()

for list_ in user_demographic_info_count.values():
    profession = list_[2]
    if profession_dict[profession] in movie_watch_count_by_profession:
        movie_watch_count_by_profession[profession_dict[profession]] += list_[4]
    else:
        movie_watch_count_by_profession[profession_dict[profession]] = list_[4]
    
print("Most active professions based on the number of ratings:")
for gender, count in movie_watch_count_by_profession.items():
    print(f"occupation {gender}: {count}")
```
