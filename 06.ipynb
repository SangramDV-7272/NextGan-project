# 6. Exploring Long Tail

### 1. Investigate the "long tail" of the dataset: How many movies receive very few ratings?

```python
with open("ml-1m\\ratings.dat","r") as f, open("ml-1m\\movies.dat","r") as d, open("ml-1m\\users.dat","r") as s : #UserID::MovieID::Rating::Timestamp
    rating_data = f.readlines()                                                                                   #MovieID::Title::Genres
    movie_data = d.readlines()                                                                                    #UserID::Gender::Age::Occupation::Zip-code
    user_data = s.readlines()
```

```python
# Count the number of ratings for each movie
movie_rating_counts = {}

for line in rating_data:

    movie_id = line.split("::")[1]
    movie_id = int(movie_id)

    if movie_id in movie_rating_counts:
        movie_rating_counts[movie_id] += 1
    else:
        movie_rating_counts[movie_id] = 1

# Find the movie with the low ratings
Min_ratings = 50
long_tail_movies = {movie_id: count for movie_id, count in movie_rating_counts.items() if count < Min_ratings}
long_tail_movies = dict(sorted(long_tail_movies.items(), key=lambda x: x[1]))


print("Movies with very few ratings:")
for movie_id, count in long_tail_movies.items():
    print(f"Movie ID {movie_id}: {count} ratings")
```

### 2. What are the characteristics of these less-rated movies compared to popular ones?

```python
movie_info = {}
for line in movie_data:
    line = line.strip()
    movie_id, movie_name, genres  = map(str,line.split('::'))
    movie_id = int(movie_id)
    genres = genres.strip().split('|')
    
    movie_info[movie_id] = [movie_name, genres]
    
movie_info[3280]
```

```python
# Analyze characteristics of less-rated movies
long_tail_movie_genres = {}
long_tail_movie_years = {}
    
movie_info[3280][1]

for movie_id in long_tail_movies:
    genres = movie_info[movie_id][1]
    for genre in genres:
        if genre in long_tail_movie_genres:
            long_tail_movie_genres[genre] += 1
        else:
            long_tail_movie_genres[genre] = 1

    year = movie_info[movie_id][0][-5:-1]
    if year in long_tail_movie_years:
        long_tail_movie_years[year] += 1
    else:
        long_tail_movie_years[year] = 1

print("\nGenres of less-rated movies:")
for genre, count in long_tail_movie_genres.items():
    print(f"{genre}: {count} movies")
```

```python
long_tail_movie_years = dict(sorted(long_tail_movie_years.items(), key=lambda x: x[0]))
print("\nYears of less-rated movies:")
for year, count in long_tail_movie_years.items():
    print(f"Year {year}: {count} movies")

print("\nless-rated movies of Year:")
less_rated_movies_of_year = dict(sorted(long_tail_movie_years.items(), key=lambda x: x[1], reverse=True)[:1])
for year, count in less_rated_movies_of_year.items():
    print(f"less-rated Year {year}: {count} ratings")
```
