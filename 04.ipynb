# Rating Distribution by Demographics

1. Investigate how ratings vary by user demographic attributes (age, gender, occupation).

```python
with open("ml-1m\\ratings.dat","r") as f, open("ml-1m\\movies.dat","r") as d, open("ml-1m\\users.dat","r") as s : #UserID::MovieID::Rating::Timestamp
    rating_data = f.readlines()                                                                                   #MovieID::Title::Genres
    movie_data = d.readlines()                                                                                    #UserID::Gender::Age::Occupation::Zip-code
    user_data = s.readlines()

profession_dict = {
                    0: "other or not specified",
                    1: "academic/educator",
                    2: "artist",
                    3: "clerical/admin",
                    4: "college/grad student",
                    5: "customer service",
                    6: "doctor/health care",
                    7: "executive/managerial",
                    8: "farmer",
                    9: "homemaker",
                    10: "K-12 student",
                    11: "lawyer",
                    12: "programmer",
                    13: "retired",
                    14: "sales/marketing",
                    15: "scientist",
                    16: "self-employed",
                    17: "technician/engineer",
                    18: "tradesman/craftsman",
                    19: "unemployed",
                    20:  "writer"}
```

```python
# this is use for collection of user demographic_info


user_demographic_info_count = dict()

for line in user_data:
    line = line.strip()
    columns = line.split('::')
    columns[2] = int(columns[2])
    columns[3] = int(columns[3])
    columns.append([0, 0])
    

    user_demographic_info_count[int(columns[0])] = columns[1:]

for line in rating_data: # 1::1193::5::978300760
    line = line.strip()
    UserID, _, Reating, _ = map(int, line.split('::'))
    
    user_demographic_info_count[UserID][4][0] += Reating
    user_demographic_info_count[UserID][4][1] += 1
    

user_demographic_info_count # 1: ['F', 1, 10, '48067', [222, 53]],

     
```

```python
# avg rating by age group
age_group_count = {
                    "Under 18": [0, 0],
                    "18-24": [0, 0],
                    "25-34": [0, 0],
                    "35-44": [0, 0],
                    "45-49": [0, 0],
                    "50-55": [0, 0],
                    "56+": [0, 0]
                }

for list_ in user_demographic_info_count.values():
    age = list_[1]
    
    if age < 18:
        age_group_count["Under 18"][0] += list_[4][0]
        age_group_count["Under 18"][1] += list_[4][1]
    elif age < 25:
        age_group_count["18-24"][0] += list_[4][0]
        age_group_count["18-24"][1] += list_[4][1]
    elif age < 35:
        age_group_count["25-34"][0] += list_[4][0]
        age_group_count["25-34"][1] += list_[4][1]
    elif age < 45:
        age_group_count["35-44"][0] += list_[4][0]
        age_group_count["35-44"][1] += list_[4][1]
    elif age < 50:
        age_group_count["45-49"][0] += list_[4][0]
        age_group_count["45-49"][1] += list_[4][1]
    elif age < 56:
        age_group_count["50-55"][0] += list_[4][0]
        age_group_count["50-55"][1] += list_[4][1]
    else:
        age_group_count["56+"][0] += list_[4][0]
        age_group_count["56+"][1] += list_[4][1]

age_group_count

print("avg rating by age group:")
for age, avg in age_group_count.items():
    print(f"Age {age}: {avg[0]/avg[1]:.2f}")
```

```python
#avg rating by gender group

movie_watch_count_gender = { i[0]: [0, 0] for i in user_demographic_info_count.values()}

for list_ in user_demographic_info_count.values():
    gender = list_[0]
    movie_watch_count_gender[gender][0] += list_[4][0]
    movie_watch_count_gender[gender][1] += list_[4][1]


movie_watch_count_gender 

print("avg rating by gender:")
for gender, avg in movie_watch_count_gender.items():
    print(f"Gender {gender}: {avg[0]/avg[1]:.2f}")
```

```python
# avg rating by occupation
 # 1: ['F', 1, 10, '48067', [222, 53]],

movie_watch_count_by_profession = {profession_dict[profession]: [0, 0] for profession in profession_dict }

for list_ in user_demographic_info_count.values():
    profession = profession_dict[list_[2]]

    movie_watch_count_by_profession[profession][0] += list_[4][0]
    movie_watch_count_by_profession[profession][1] += list_[4][1]
    
movie_watch_count_by_profession
print("avg rating by occupation:")
for gender, avg in movie_watch_count_by_profession.items():
    print(f"occupation {gender}: {avg[0]/avg[1]:.2f}")
```

2. Are there specific genres preferred by certain age groups or occupations?

```python
MovieID_with_Genre = dict()
for line in movie_data:
    userID, _, Genres = map(str, line.split("::"))
    userID = int(userID)
    Genres = Genres.strip()
    Genre =list(map(str, Genres.split("|")))

    MovieID_with_Genre[userID] = Genre

MovieID_with_Genre
```

```python
# Count the number of ratings for each genre by age group
genre_rating_count_by_age = dict()

def fun_(age, genres):
    global genre_rating_count_by_age
    if age not in genre_rating_count_by_age:
        genre_rating_count_by_age[age] = dict()
    for genre in genres:
        if genre not in genre_rating_count_by_age[age]:
            genre_rating_count_by_age[age][genre] = 1
        else:
            genre_rating_count_by_age[age][genre] += 1

for age in rating_data:# 1::1193::5::978300760
    userID, MovieID, _, _ = map(int, age.split('::'))
    age = user_demographic_info_count[userID][1]
    genres = MovieID_with_Genre[MovieID]

    if age < 18:
        fun_("Under 18", genres)

    elif age < 25:
        fun_("18-24", genres)
        
    elif age < 35:
        fun_("25-34", genres)
        
    elif age < 45:
        fun_("35-44", genres)
        
    elif age < 50:
        fun_("45-49", genres)
        
    elif age < 56:
        fun_("50-55", genres)
        
    else:
        fun_("56+", genres)

print("Genre preferences by age group:")
for age, genre in genre_rating_count_by_age.items():
    print(f"Age {age} :{max(genre, key=genre.get)}({genre[max(genre, key=genre.get)]})")


    
```

```python
genre_rating_counts_by_profession = dict()

for line in rating_data:
    userID, MovieID, _, _ = map(int, line.split('::'))
    
    profession = profession_dict[user_demographic_info_count[userID][2]]
    genres = MovieID_with_Genre[MovieID]

    for genre in genres:
        if profession not in genre_rating_counts_by_profession:
            genre_rating_counts_by_profession[profession] = dict()

        if genre not in genre_rating_counts_by_profession[profession]:
            genre_rating_counts_by_profession[profession][genre] = 1

        else:
            genre_rating_counts_by_profession[profession][genre] += 1



print("Genre preferences by profession:")
for profession, genre in genre_rating_counts_by_profession.items():
    print(f"profession {profession} :{max(genre, key=genre.get)}({genre[max(genre, key=genre.get)]})")
```
